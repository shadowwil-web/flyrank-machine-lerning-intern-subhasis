## 5. Why ML beats a fixed rule here

A fixed rule (like a hand-coded baseline) treats all pages that cross a hard threshold exactly the same. If a rule flags pages with impressions > 500 and trend == down, it cannot differentiate between a massive flagship page losing 40% of its traffic and a medium page losing 5%. Machine learning can capture the non-linear relationships between volume, age, position, and engagement, allowing for a much more nuanced risk score.

Instead of a human guessing how much mathematical weight to give "freshness" versus "position drop," a machine learning model learns the true weights directly from historical outcomes. Furthermore, ML outputs a precise probability score, allowing us to strictly rank the review queue from highest to lowest risk, rather than just dumping pages into a generic "needs review" bucket. The starter dataset already proves this: the baseline fixed rule achieved a Precision@50 of only 0.240, while the Random Forest model achieved 0.740.

*What makes the pattern too messy for an if-statement?*

Content decay is not a simple, linear process. An if statement forces us to use rigid, arbitrary thresholds (e.g., if impressions_90d > 500 and ctr < 0.5). This is far too messy for hand-coded rules because a fixed threshold completely ignores a critical page that might have 499 impressions and a terrible 0.1 CTR. Search visibility depends on complex, non-linear interactions between content age, search volume, position drops, and engagement rates that simple rules cannot handle simultaneously.
