## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**The Decision:** Which specific web pages are showing signs of structural or visibility decline and require human intervention?

**The Grain** (Unit of Analysis): A single pseudonymized content item (page).

**The Action:** A content strategy team member will review the top-ranked pages in the generated queue and update, rewrite, or expand the material to fix the flagged issue.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# 1. Load the starter dataset directly from your public GitHub using the Raw URL
url = 'https://raw.githubusercontent.com/shadowwil-web/flyrank-machine-lerning-intern-subhasis/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# 2. Filter data using the same rules as the pipeline
# Keep only pages that have demand and a known history window
active_df = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)].copy()

# Calculate numbers for framing evidence
total_pages = len(active_df)
declining_pages = len(active_df[active_df['trend_direction'] == 'down'])
pct_declining = (declining_pages / total_pages) * 100

total_impressions = active_df['impressions_90d'].sum()
declining_impressions = active_df[active_df['trend_direction'] == 'down']['impressions_90d'].sum()
pct_traffic_at_risk = (declining_impressions / total_impressions) * 100

print(f"--- LANE 2 DATA EVIDENCE ---")
print(f"1. Total active pages analyzed: {total_pages:,}")
print(f"2. Pages actively declining (Proxy Target): {declining_pages:,} ({pct_declining:.2f}%)")
print(f"3. Share of total search impressions currently at risk: {pct_traffic_at_risk:.2f}%")
```

What I Can Claim:-
